```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Data
X = np.array([[10], [20], [30], [40], [50]])  # Jumlah pelanggan
y = np.array([200, 400, 600, 800, 1000])      # Pendapatan dalam ribu rupiah

# Buat model regresi linier
model = LinearRegression()
model.fit(X, y)

# Hasil model
slope = model.coef_[0]
intercept = model.intercept_
r_squared = model.score(X, y)

print(f"Persamaan regresi: Y = {slope:.2f}X + {intercept:.2f}")
print(f"R-squared: {r_squared:.2f}")

# Prediksi
X_pred = np.array([[35]])  #Ditanya prediksi pendapatan 35 pelangan
y_pred = model.predict(X_pred)
print(f"Prediksi pendapatan untuk 35 pelanggan: {y_pred[0]:.2f} ribu rupiah")

# Visualisasi
plt.scatter(X, y, color='orange', label='Data asli')
plt.plot(X, model.predict(X), color='blue', label='Regresi linier')
plt.xlabel('Jumlah Pelanggan')
plt.ylabel('Pendapatan (ribu rupiah)')
plt.title('Regresi Linier Jumlah Pelanggan vs Pendapatan')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Data
X = np.array([[1], [3], [5], [7], [9]])  # Usia kendaraan (tahun)
y = np.array([150, 120, 90, 60, 30])     # Harga jual (juta)

# Membuat model regresi linier
model = LinearRegression()
model.fit(X, y)

# Mengambil koefisien (slope) dan intercept
slope = model.coef_[0]
intercept = model.intercept_

print(f"Persamaan regresi: Y = {intercept:.2f} + ({slope:.2f})X")

# Prediksi harga jual untuk usia 4 tahun
usia_prediksi = np.array([[4]])
harga_prediksi = model.predict(usia_prediksi)

print(f"Prediksi harga jual untuk usia 4 tahun: {harga_prediksi[0]:.2f} juta rupiah")
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Data
X = np.array([
    [50, 2],
    [60, 3],
    [70, 4],
    [80, 5],
    [90, 6]
])  # X1 = pengunjung, X2 = menu baru

y = np.array([400, 480, 560, 640, 720])  # Penjualan (ribu rupiah)

# Model regresi linier berganda
model = LinearRegression()
model.fit(X, y)

# Koefisien regresi
intercept = model.intercept_
b1, b2 = model.coef_

print(f"Model regresi: Y = {intercept:.2f} + ({b1:.2f})X1 + ({b2:.2f})X2")

# Prediksi jika pengunjung = 75 dan menu baru = 4
X_pred = np.array([[75, 4]])
y_pred = model.predict(X_pred)

print(f"Prediksi penjualan (pengunjung=75, menu baru=4): {y_pred[0]:.2f} ribu rupiah")
```
